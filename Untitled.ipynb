```python
from main import *
import matplotlib.pyplot as plt

df = data_transform(data_in())
df = df.drop(72)
df = df.drop(544)
```

```python
def graph(focus, variables):
    fig, ax = plt.subplots(3, 2,
                           figsize=(20, 10))
    fig.suptitle(focus.name)
    i = 0
    for var in variables:
        if i>2:
            ax[i-3][0].scatter(focus, var)
            ax[i-3][0].set_title(var.name)
        else:
            ax[i][1].scatter(focus, var)
            ax[i][1].set_title(var.name)
        i += 1
```

```python
graph(df['Total Vineyard Area'],
        [df['Grapes harvested'],
        df['Total water used'],
        df['total irrigation'],
        df['total fuel'],
        df['#No. Passes'],
        df['Applied Fertiliser'],
      ]
     )
```

```python
graph(df['Grapes harvested'],
        [
            df['Total Vineyard Area'],
            df['Total water used'],
            df['total irrigation'],
            df['total fuel'],
            df['#No. Passes'],
            df['Applied Fertiliser'],
      ]
     )
```

```python
graph(df['Total water used'],
        [
            df['Total Vineyard Area'],
            df['Grapes harvested'],
            df['total irrigation'],
            df['total fuel'],
            df['#No. Passes'],
            df['Applied Fertiliser'],
      ]
     )
```

```python
graph(df['total fuel'],
        [
            df['Total Vineyard Area'],
            df['Grapes harvested'],
            df['Total water used'],
            df['total irrigation'],
            df['#No. Passes'],
            df['Applied Fertiliser'],
      ]
     )
```

```python
graph(df['#No. Passes'],
        [
            df['Total Vineyard Area'],
            df['Grapes harvested'],
            df['Total water used'],
            df['total irrigation'],
            df['total fuel'],
            df['Applied Fertiliser'],
      ]
     )
```

```python
graph(df['Applied Fertiliser'],
        [
            df['Total Vineyard Area'],
            df['Grapes harvested'],
            df['Total water used'],
            df['total irrigation'],
            df['total fuel'],
            df['#No. Passes'],
      ]
     )
```
